```python
# Mini Project-09: EDA on Wine Quality Dataset (UCI)

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load the dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
df = pd.read_csv(url, sep=';')

# Step 2: Basic Exploration
print("\nüìù Dataset Shape:", df.shape)
print("\nüìÑ First 5 Rows:")
print(df.head())

print("\nüß± Column Info:")
df.info()

print("\nüìä Summary Stats:")
print(df.describe())

# Step 3: Check for missing values
print("\nüîç Missing Values:")
print(df.isnull().sum())

# Step 4: Distribution of wine quality
plt.figure(figsize=(7, 4))
sns.countplot(x='quality', data=df, palette='Set2')
plt.title("Distribution of Wine Quality Ratings")
plt.xlabel("Quality")
plt.ylabel("Count")
plt.show()

# Step 5: Correlation Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()

# Step 6: Boxplots to detect outliers
plt.figure(figsize=(15, 6))
sns.boxplot(data=df, palette='Pastel1')
plt.xticks(rotation=45)
plt.title("Boxplot of All Features")
plt.show()

# Step 7: Pairplot (Optional: takes time)
# sns.pairplot(df, hue='quality', palette='husl')

# Step 8: Quality vs Alcohol (example analysis)
plt.figure(figsize=(8,5))
sns.boxplot(x='quality', y='alcohol', data=df, palette='Set3')
plt.title("Alcohol Content vs Wine Quality")
plt.show()

# Optional: Save to CSV or export filtered data
# df.to_csv("winequality_cleaned.csv", index=False)
```
